# Notebook 05: Explainability - SHAP & LIME Analysis

Provides model interpretability using:
- **SHAP**: Global and local feature importance for all models
- **LIME**: Local explanation of individual predictions
- **Cross-model feature importance comparison**

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

from src.feature_engineering import load_prepared_data, prepare_features
from src.ml_models import load_model, predict_with_model
from src.explainability import (
    full_shap_analysis, full_lime_analysis,
    cross_model_feature_importance
)
```

## 5.1 Load Data & Models

```python
FORMAT_KEY = 'mens_odi'
train_df, test_df = load_prepared_data(FORMAT_KEY)

X_train, y_train, feature_names = prepare_features(train_df)
X_test, y_test, _ = prepare_features(test_df)

# Load trained models
models = {}
for name in ['xgboost', 'random_forest', 'lightgbm', 'neural_network']:
    try:
        model, scaler = load_model(name, format_key=FORMAT_KEY)
        models[name] = {'model': model, 'scaler': scaler}
        print(f"Loaded: {name}")
    except FileNotFoundError:
        print(f"Not found: {name}")

print(f"\nFeatures: {feature_names}")
print(f"Test samples: {len(X_test)}")
```

## 5.2 SHAP Analysis - XGBoost

```python
if 'xgboost' in models:
    xgb_shap, xgb_explainer = full_shap_analysis(
        models['xgboost']['model'], X_test,
        model_name='XGBoost',
        model_type='tree',
        format_key=FORMAT_KEY,
        sample_size=1000,
    )
    print("XGBoost SHAP analysis complete")
```

## 5.3 SHAP Analysis - Random Forest

```python
if 'random_forest' in models:
    rf_shap, rf_explainer = full_shap_analysis(
        models['random_forest']['model'], X_test,
        model_name='RandomForest',
        model_type='tree',
        format_key=FORMAT_KEY,
        sample_size=1000,
    )
    print("Random Forest SHAP analysis complete")
```

## 5.4 SHAP Analysis - LightGBM

```python
if 'lightgbm' in models:
    lgb_shap, lgb_explainer = full_shap_analysis(
        models['lightgbm']['model'], X_test,
        model_name='LightGBM',
        model_type='tree',
        format_key=FORMAT_KEY,
        sample_size=1000,
    )
    print("LightGBM SHAP analysis complete")
```

## 5.5 SHAP Analysis - Neural Network

Uses KernelExplainer (slower, but works for any model).

```python
if 'neural_network' in models:
    # NN SHAP is slower - use smaller sample
    nn_shap, nn_explainer = full_shap_analysis(
        models['neural_network']['model'], X_test,
        model_name='NeuralNetwork',
        model_type='kernel',
        scaler=models['neural_network']['scaler'],
        format_key=FORMAT_KEY,
        sample_size=200,  # Smaller for KernelExplainer
    )
    print("Neural Network SHAP analysis complete")
```

## 5.6 Cross-Model Feature Importance

```python
# Collect all SHAP values
shap_results = {}
if 'xgboost' in models:
    shap_results['XGBoost'] = xgb_shap
if 'random_forest' in models:
    shap_results['RandomForest'] = rf_shap
if 'lightgbm' in models:
    shap_results['LightGBM'] = lgb_shap
if 'neural_network' in models:
    shap_results['NeuralNetwork'] = nn_shap

if shap_results:
    importance_df = cross_model_feature_importance(
        shap_results, feature_names,
        format_key=FORMAT_KEY,
        top_n=15
    )
    print("\nCross-model feature importance:")
    print(importance_df)
```

## 5.7 LIME Analysis

```python
# LIME for XGBoost (best tree model typically)
if 'xgboost' in models:
    xgb_lime = full_lime_analysis(
        models['xgboost']['model'],
        X_train, X_test, y_test,
        model_name='XGBoost',
        format_key=FORMAT_KEY,
        n_instances=5,
    )
    print("XGBoost LIME analysis complete")
```

```python
# LIME for Neural Network
if 'neural_network' in models:
    nn_lime = full_lime_analysis(
        models['neural_network']['model'],
        X_train, X_test, y_test,
        model_name='NeuralNetwork',
        scaler=models['neural_network']['scaler'],
        format_key=FORMAT_KEY,
        n_instances=5,
    )
    print("Neural Network LIME analysis complete")
```
